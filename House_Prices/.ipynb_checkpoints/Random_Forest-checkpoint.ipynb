```python
import numpy as np
import pandas as pd
import sklearn as sk
import sklearn.ensemble
np.random.seed(0)
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
df = pd.concat([train, test], ignore_index=True)
df.drop(["Id", "SalePrice"], axis=1, inplace=True)
df = pd.get_dummies(df)
df = df.fillna(df.mean())
```

```python
df.dtypes
df.head(30)
```

```python
#df.shape
```

```python
x = df[:train.shape[0]]
y = train.SalePrice
x_test = df[train.shape[0]:]
```

```python
model_rf = sk.ensemble.RandomForestRegressor(n_estimators=1000)
model_rf.fit(x, y)
```

```python
y_test = model_rf.predict(x_test)
```

```python
solution = pd.DataFrame({"Id": test.Id, "SalePrice": y_test})
solution.to_csv("rf.csv", index = False)
```
